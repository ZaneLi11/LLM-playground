```python
from langchain_ollama import OllamaLLM

model = OllamaLLM(model = "deepseek-r1:8b")

response = model.invoke(input="who are u?")

print(response)
```

```python
import os
from dotenv import load_dotenv

from langchain_google_genai import ChatGoogleGenerativeAI

load_dotenv()
gemini_api_key = os.getenv("GEMINI_API_KEY")

if not gemini_api_key:
    raise ValueError("Can't find GEMINI_API_KEY")

model = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash", 
    google_api_key=gemini_api_key
)

response = model.invoke("who are u?")


print(response.content)
```

```python
model = OllamaLLM(model = "deepseek-r1:8b")

response = model.stream(input="Could you list Top 10 research universities around the world?")

for chunk in response:
    print(chunk, end = "", flush=True)
```

```python
model = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash", 
    google_api_key=gemini_api_key
)

response = model.stream("Could you list Top 10 research universities in US?")


for chunk in response:
    print(chunk.content, end="", flush=True)
```

```python
from langchain_core.messages import HumanMessage, SystemMessage, AIMessage
```

```python
model = OllamaLLM(model = "deepseek-r1:8b")

messages = [
    SystemMessage(content="你是一个暴躁老师，你喜欢使用简短的语句和狂躁的语气。"),
    HumanMessage(content="如何快速学好Python编程？"),
    AIMessage(content="别做梦了！想快速学好Python？做梦！基础语法先啃透，变量、循环、函数、类，一个都不能少！代码敲烂了再问，别问能不能速成!做项目，看源码，debug到吐血，这才是正道!别指望一口吃成胖子，编程是爬山，慢慢往上爬！"),
    HumanMessage(content="那你能告诉我正确的学习路径吗？")
]

for chunk in model.stream(input=messages):
    print(chunk, end = "", flush=True)
```
