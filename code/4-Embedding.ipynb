```python
import os
from dotenv import load_dotenv
from langchain_google_genai import GoogleGenerativeAIEmbeddings

load_dotenv()
gemini_api_key = os.getenv("GEMINI_API_KEY")

if not gemini_api_key:
    raise ValueError("Cannot find GEMINI_API_KEY in .env file.")

# Initialize the embedding model using the most stable v1 model
# "models/embedding-001" is universally supported across all API endpoints
embeddings_model = GoogleGenerativeAIEmbeddings(
    model="models/gemini-embedding-001", 
    google_api_key=gemini_api_key
)

print("Calling Google Embedding Model...\n")
```

```python
print(embeddings_model.embed_query("i love u"))
```

```python
from langchain_ollama import OllamaEmbeddings


# Initialize the embedding model using the most stable v1 model
# "models/embedding-001" is universally supported across all API endpoints
embeddings_model = OllamaEmbeddings(
    model="nomic-embed-text"
)

print("Calling Local Embedding Model...\n")
```

```python
print(embeddings_model.embed_query("i love u"))
print(embeddings_model.embed_documents(["how are u?","who are u?","what's ur name?"]))
```
