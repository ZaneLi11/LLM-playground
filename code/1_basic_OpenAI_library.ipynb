```python
import os
from dotenv import load_dotenv
from openai import OpenAI

# read .env
load_dotenv()
gemine_api  = os.getenv("GEMINI_API_KEY")

# Initialize the OpenAI client
client = OpenAI(
    api_key=gemine_api,
    base_url="https://generativelanguage.googleapis.com/v1beta/openai/"
)

# Create a chat completion request to the Gemini model
response = client.chat.completions.create(
    model = "gemini-2.5-flash",
    messages = [
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": "can u help me find out all US news Top 10 universities"}
    ]
)

print(response.choices[0].message.content)
```

```python
# Create a chat completion request to the Gemini model
response = client.chat.completions.create(
    model = "gemini-2.5-flash",
    messages = [
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": "I have 3 red dolls."},
        {"role": "assistant", "content": "Glad to hear that"},
        {"role": "user", "content": "I have 6 green dolls."},
        {"role": "assistant", "content": "it's so nice!"},
        {"role": "user", "content": "How many dolls do I have in taotal?"}  # Question based on history information
    ],
    stream = True
)

for chunk in response:
    print(chunk.choices[0].delta.content,end = " ", flush = True)
```
