```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_ollama import ChatOllama
from langchain_core.output_parsers import StrOutputParser, JsonOutputParser
from langchain_core.runnables import RunnableLambda

parser_str = StrOutputParser()
parser_json = JsonOutputParser()

deepseek = ChatOllama(
    model="deepseek_r1:8b"
)

Prompt_template1 = ChatPromptTemplate.from_template(
    "My Chinese is {Chinese_name}. Can you select an English name for me? Please generate one name without extra information!"
    )

Prompt_template2 = ChatPromptTemplate.from_template(
    "My Englsih name is {Name}, can you explain this name shortly?"
)

deepseek = ChatOllama(
    model="deepseek-r1:8b"
)

# my_func = RunnableLambda(lambda ai_msg:dict({"Name":ai_msg.content}))

def func(ai_msg):
    dict = {"Name":ai_msg.content} 
    print(ai_msg.content)
    return dict

my_func = RunnableLambda(func)


Chinese_name = "李子正"

chain  = Prompt_template1 | deepseek | my_func | Prompt_template2 | deepseek | parser_str

answer = chain.stream({"Chinese_name":Chinese_name})

for chunk in answer:
    print(chunk, end = "", flush=True)
```
