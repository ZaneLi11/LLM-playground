```python
from langchain_core.output_parsers import StrOutputParser
from langchain_ollama import OllamaEmbeddings
from langchain_chroma import Chroma
from langchain_ollama import ChatOllama
from langchain_core.prompts import ChatPromptTemplate
```

```python
deepseek = ChatOllama(
    model="deepseek-r1:8b"
)
embeddings = OllamaEmbeddings(model="nomic-embed-text")
vector_store = Chroma(
    collection_name="3D_disaster_housing",
    embedding_function=embeddings,
    persist_directory="./Vectorstores"
)
```

```python
Template = ChatPromptTemplate.from_messages(
    [
        ("system", "base on the information here: {context}, answer user's questions."),
        ("user","{input}")
    ]
)
```

```python
query_text = "please tell me someting about 3D printing housing."
results = vector_store.similarity_search(query_text, k=3)
```

```python
reference_content = "["
for chunk in results:
    reference_content += chunk.page_content
reference_content += "]"
```

```python
def print_prompt(prompt):
    print(prompt.to_string(),"\n")
    return prompt
```

```python
chain = Template|print_prompt|deepseek|StrOutputParser()
```

```python
res = chain.invoke({"input":query_text,"context":reference_content})
```
