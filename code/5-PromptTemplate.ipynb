```python
from langchain_core.prompts import PromptTemplate
from langchain_ollama import OllamaLLM


template_1 = PromptTemplate.from_template(
    "我现在正在进行{programming_language}语言编程，我想完成一个{goal}，请用文字简要说明需要实现的类和功能以及容易出现逻辑错误的地方。"
)

prompt_1 = template_1.format(programming_language = "Python",goal = "贪吃蛇小游戏")

print(prompt_1)

model_1 = OllamaLLM(
    model="deepseek-r1:8b"
)

response_1 = model_1.stream(prompt_1)

for chunk in response_1:
    print(chunk,end="",flush=False)
```

```python
chain = template_1 | model_1

response_2 = chain.stream(input={
    "programming_language" : "Python",
    "goal" : "石头剪刀布"
})

for chunk in response_2:
    print(chunk,end="",flush=False)
```

```python
from langchain_core.prompts import FewShotPromptTemplate , PromptTemplate

example_template = PromptTemplate.from_template("word:{word},antonym:{antonym}")

examples_data = [
    {"word": "fast","antonym":"slow"},
    {"word":"big","antonym":"small"}
]

Fewshot_templates = FewShotPromptTemplate(
    example_prompt = example_template,
    examples = examples_data,
    prefix = "Tell me the antonym of the word. Here is examples:",
    suffix = "Base on examples, tell me antonyms fror {input_1} and {input_2}",
    input_variables = ["input_1","input_2"]
)

```

```python
response_3 = model_1.invoke(input=Fewshot_templates.invoke(input={"input_1":"tall","input_2":"ugly"}))
```

```python
print(response_3)
```

```python
print(Fewshot_templates.invoke({"input_1":"tall","input_2":"ugly"}))
print(type(Fewshot_templates.invoke({"input_1":"tall","input_2":"ugly"})))
```
