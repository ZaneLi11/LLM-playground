```python
from langchain_community.document_loaders import CSVLoader

loader = CSVLoader(
    file_path="./data/people.csv",
    csv_args={
        "deliminater":",",
        "quotechar":'"',
        # "filednames":["a","b","c","d"]
        # apply if data doesn't have col names
    },
    # encoding="utf-8"
)

documents = loader.load()

for document in documents:
    print(type(document),document,"\n")
```

```python
documents = loader.lazy_load()

for document in documents:
    print(document)
```
